Coluna com dados faltantes: Data Abertura

Tipo de dado: String

Estamos lidando com uma coluna de data, por isso, me parece estranho usar medidas de posição, como média e mediana, para preencher os valores ausentes.

Uma maneira possível de encontrar um padrão na data de abertura (da licitação) é pensar no tempo médio entre abertura da licitação e homologação da compra. Pensando nisso, vou usar essa informação para preencher os valores faltantes.

```python
# Transformando colunas de data em tipo data:

licitacoes['Abertura_como_data'] = pd.to_datetime(licitacoes['Data Abertura'],dayfirst=True,format='%d/%m/%Y')
licitacoes['Resultado_compra_como_data'] = pd.to_datetime(licitacoes['Data Resultado Compra'],dayfirst=True,format='%d/%m/%Y')
```

Com as colunas no tipo certo, vou criar uma nova coluna indicando quanto tempo de existência cada empresa contratada tinha quando a contratação foi homologada no Diário Oficial da União (segundo o [Dicionário de Dados](https://portaldatransparencia.gov.br/pagina-interna/603389-dicionario-de-dados-licitacoes) do governo):

```python
licitacoes['Duracao_licitacao'] =  (licitacoes['Resultado_compra_como_data'] - licitacoes['Abertura_como_data']).dt.days
licitacoes['Duracao_licitacao']
```

```python
licitacoes['Duracao_licitacao'].describe()
```

Existe bastante variação na coluna criada. Chamam a atenção o mínimo (uma licitação finalizada no mesmo dia que foi aberta?) e no máximo (mais de 4 anos para finalizar uma licitação?), mas nenhum dos casos é impossível no cenário brasileiro. Somando a isso o fato de que não temos muito mais informações sobre o dataset, vamos seguir com esses valores.

```python
# Transformando a coluna Duracao_licitacao em duração para poder fazer a operação com uma coluna de data

licitacoes['Duracao_licitacao'] =  (licitacoes['Resultado_compra_como_data'] - licitacoes['Abertura_como_data'])
```

```python
# Acrescentando a média de duração nas linhas com valor vazio na coluna nova:
mediana_duracao = licitacoes['Duracao_licitacao'].median()
licitacoes['Duracao_licitacao'] = licitacoes['Duracao_licitacao'].fillna(mediana_duracao)

# dados['IDADE'] = dados[dados['FAIXA IDADE'] == '17-21']['IDADE'].fillna(media17)
```

```python
# Preenchendo as linhas vazias da coluna 'Abertura_como_data' a partir da nova coluna

licitacoes.loc[(licitacoes['Abertura_como_data'].isnull()),'Abertura_como_data'] = licitacoes['Resultado_compra_como_data']-licitacoes['Duracao_licitacao']
```

```python
# Colocando as informações na coluna original e no tipo original de dado:

licitacoes['Data Abertura'] = licitacoes['Abertura_como_data'].astype('str')
```

```python
# Verificando se deu certo:
licitacoes.info()
```

```python
# Apagando as colunas criadas pra deixar o DF mais limpo:
licitacoes = licitacoes.drop(['Abertura_como_data','Resultado_compra_como_data','Duracao_licitacao'], axis=1)
```

## Tratando valores discrepantes (Outliers)

```python
lista_idades = [26, 30, 32, 22, 26, 35, 400, 20, 43, 31, 23]
```

```python
# Calculando desvio padrão

desvio = np.std(lista_idades)
desvio
```

### Identificando valores discrepantes

#### Método 1: a partir da mediana

Achando o limite superior:

```python
media = np.mean(lista_idades)
media + 3 * desvio
```

Achando o limite inferior:

```python
media - 3 * desvio
```

#### Método 2: a partir dos quartis

Porém, como estamos trabalhando com idade, esses limites não fazem sentido.

Então, vamos usar mediana e quartis, em vez deles. O primeiro quartil é a metade debaixo da mediana; o segundo, a metade de cima.

Vamos visualizar isso com um boxplot:

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(lista_idades)
```

A linha laranja é a mediana.

A borda de cima da caixa, o quartil Q3.

A boroda de cima, o quartil Q1.

As linhas acima e abaixo da caixa, os limites superior e inferior.

```python
# Verificando outliers na coluna de salário:

plt.boxplot(dados['SALARIO'])
```

```python
q1 = dados['SALARIO'].quantile(0.25)
q1
```

```python
q3 = dados['SALARIO'].quantile(0.75)
q3
```

```python
# Interquartil = Intervalo entre o Q1 e o Q3
iqr = q3 - q1
iqr
```

```python
# Limite superior: q3 + 1.5 interquartil
# Limite inferior: q1 - 1.5 interquartil

lim_superior = 3 + (1.5 * iqr)
lim_inferior = q1 - (1.5 * iqr)

print(lim_superior)
print(lim_inferior)
```

Como estamos lidando com salários, podemos considerar que o limite inferior é 0.

```python
dados['FAIXA SALARIAL'].value_counts()
```

Nosso limite superior foi de 10567.5. Porém, olhando os respondentes por faixa salarial, vemos que quem ganha mais do que isso não é tão outlier assim (por exemplo, o 4º grupo com mais respondentes foi de 12 a 16 mil). Então, vamos optar por outro método de identificação de outliers.

#### Método 3: distância da média

Usamos como critério a distância que um determinado dado está da média, usando o desvio padrão para encontrar essa distância.

```python
media_salario = dados['SALARIO'].mean()
media_salario
```

```python
desvio_salario = dados['SALARIO'].std()
desvio_salario
```

Nesse caso, o limite superior será a média + um certo número de desvios padrões.

Esse número pode valiar segundo a sensibilidade da análise e da distribuição dos dados. O que determina isso é a análise de contexto.

Nesse caso, como vimos no boxplot que os outliers estão bastante longe dos valores comuns, vamos usar 3 desvios padrão.

```python
lim_superior3 = media_salario + (3 * desvio_salario)
lim_superior3
```

### Lidando com outliers

- Um caminho é removê-los, mas isso pode distorcer a distribuição dos dados e influenciar as conclusões.
- Outro caminho é substituí-los por valores mais adequados:
  - Mediana (em vez da média, que é sensível a valores discrepantes)
  - Valor máximo (considerando outras variáveis do dataset)
- Em outros cenários, vai ser necessário manter os valores, pois são eventos reais e importantes para a análise.

No nosso caso, vamos usar a coluna de faixa salarial para lidar com o outlier:

```python
dados[dados['SALARIO']>lim_superior3]['FAIXA SALARIAL'].value_counts()
```

```python
# Podemos fazer a média de salários entre 30k e 40k que não sejam outliers e substituir nos outliers, depois fazer o mesmo no outro grupo:

media_30k = dados[(dados['FAIXA SALARIAL'] == 'de R$ 30.001/mês a R$ 40.000/mês') & (dados['SALARIO']<lim_superior3)]['SALARIO'].mean()
media_30k
```

```python
dados.loc[(dados['FAIXA SALARIAL'] == 'de R$ 30.001/mês a R$ 40.000/mês') & (dados['SALARIO']>=lim_superior3),'SALARIO'] = media_30k
```

```python
# Fazendo o mesmo com valores acima de 40k:

media_40k = dados[(dados['FAIXA SALARIAL'] == 'Acima de R$ 40.001/mês') & (dados['SALARIO'] < lim_superior3)]['SALARIO'].mean()
media_40k
```

```python
dados.loc[(dados['FAIXA SALARIAL'] == 'Acima de R$ 40.001/mês') & (dados['SALARIO'] >= lim_superior3),'SALARIO'] = media_40k
```

```python
dados[dados['SALARIO']>lim_superior3]['FAIXA SALARIAL'].value_counts()
```

## Intervalo de confiança e distribuição amostral

```python
salarios = dados['SALARIO']
salarios
```

```python
# Vamos usar numpy, já que agora estramos trabalhando com uma lista
media_amostral = np.mean(salarios)
media_amostral
```

```python
# Calculando desvio padrão:
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
# Criando uma variável de nível de confiaça:
nivel_confianca = 0.95
```

```python
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
# Erro padrão é a diferença de médias de diversas amostras
# Para calculá-lo, vamos usar uma nova biblioteca, que divide nossa amostra em grupos de amostra

from scipy import stats
```

```python
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
# Calculando intervalo de confiança
# Parâmetros:
#   nível de confiança,
#   nº de graus de liberdade da distribuição (tamanho da amostra - 1),
#   loc (média amostral),
#   scale (erro padrão da média)

intervalo_confianca = stats.t.interval(nivel_confianca,tamanho_amostra-1,loc=media_amostral,scale=erro_padrao)
intervalo_confianca
```

O resultado "(9655.186295869078, 10153.593923757535)" significa que podemos afirmar com 95% de confiança que a média salarial do Cientista de Dados no Brasil é entre 9.655 a 10.153 reais.

#### Exercício - Intervalo de confiança

Calcular, com 90% de confiança, a média salarial por gênero.

##### Por cargo:

```python
# Não consegui rodar usando o erro amostral da amostra filtrada por gênero, porque ficava pequena demais.
# Então, fiz com o erro padrão da amostra completa.

media_salario_masculino = dados[(dados['GENERO'] == 'Masculino')]['SALARIO'].mean()
desvio_salario_masculino = dados[(dados['GENERO'] == 'Masculino')]['SALARIO'].std()
nivel_confianca = 0.8
tamanho_amostra_salario_masculino = len(dados[(dados['GENERO'] == 'Masculino')]['SALARIO'])
erro_padrao = stats.sem(salarios)
intervalo_confianca = stats.t.interval(nivel_confianca,tamanho_amostra_salario_masculino-1,loc=media_salario_masculino,scale=erro_padrao)
print(f'Média salariaL de homens, com {nivel_confianca * 100}% de confiança, é de {intervalo_confianca[0]} a {intervalo_confianca[1]} reais.')
```

```python
# Não consegui rodar usando o erro amostral da amostra filtrada por gênero, porque ficava pequena demais.
# Então, fiz com o erro padrão da amostra completa.

media_salario_feminino = dados[(dados['GENERO'] == 'Feminino')]['SALARIO'].mean()
desvio_salario_feminino = dados[(dados['GENERO'] == 'Feminino')]['SALARIO'].std()
nivel_confianca = 0.8
tamanho_amostra_salario_feminino = len(dados[(dados['GENERO'] == 'Feminino')]['SALARIO'])
erro_padrao = stats.sem(salarios)
intervalo_confianca = stats.t.interval(nivel_confianca,tamanho_amostra_salario_feminino-1,loc=media_salario_feminino,scale=erro_padrao)
print(f'Média salarial de mulheres, com {nivel_confianca * 100}% de confiança, é de {intervalo_confianca[0]} a {intervalo_confianca[1]} reais.')
```
